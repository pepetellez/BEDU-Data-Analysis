```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
ax = sns.distplot(df['glucose'], kde=False, norm_hist=False)
ax.set(title="Glucosa", xlabel="Concetración de Glucosa", ylabel="Conteo")
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
ax = sns.barplot(pregnancy_counts.index, pregnancy_counts)
ax.set(title="Embarazos", xlabel="Número de embarazos", ylabel="Conteo")
```

### c)

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
ax = sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True},  label="Sanos")
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True}, label="Enfermos", ax=ax)
ax.set_title('Distribuciones de Glucosa de personas según su categoría', fontsize=13, pad=15);
ax.set(xlabel='Concentración de Glucosa', ylabel="Densidad");
ax.legend(loc='upper right');
```

### d)

```python
fig = plt.figure(figsize=(10,10))
ax = sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True}, label="Presión Sanguínea")
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True}, label="Glucosa")
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True}, label="Insulina")
ax.set_title('Distribución de mediciones de personas enfermas ', fontsize=20, pad=20);
ax.set_xlabel('Valores', fontsize=15);
ax.set_ylabel('Densidad', fontsize=15);
ax.legend(loc='upper right');
```
